```python
from pinballrt.dust import Dust
from astropy.modeling import models
import astropy.units as u
import numpy as np

import matplotlib.pyplot as plt
```

```python
data = np.loadtxt(os.path.join(os.path.dirname(__file__), "data/dustkappa_yso.inp"), skiprows=2)

lam = data[::-1,0].copy() * u.micron
kabs = data[::-1,1].copy() * u.cm**2 / u.g
ksca = data[::-1,2].copy() * u.cm**2 / u.g

d = Dust(lam, kabs, ksca)
```

```python
d.learn(model="random_nu", hidden_units=3*(48,))
```

```python
d.fit(epochs=100)
```

```python
T = np.repeat(10.**np.random.uniform(-1., 4., 1), 10000)

nu = d.random_nu_manual(T)
nu2 = d.random_nu_ml(T)

counts, bins, _ = plt.hist(nu, 100)
plt.hist(nu2, 100)

bb = models.BlackBody(temperature=T[0] * u.K)
pdf = bb(bins[1:]*u.GHz).value * d.interpolate_kabs(bins[1:]*u.GHz)
pdf *= counts.max() / pdf.max()

plt.plot(bins[1:], pdf, '-')

plt.show()
```
